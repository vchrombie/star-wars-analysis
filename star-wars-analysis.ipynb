# Star Wars Analysis

Lets start with collecting the data and loading it into a dataframe using pandas.

```python
import pandas as pd

swdf = pd.read_csv('https://query.data.world/s/lvvugqww5m5lu55mlpgabdwxmtykwc', encoding="ISO-8859-1")
```

Now, it is clear that only the ResponseID is in the required format. Lets start cleaning the data and formatting the data according to our use.

Step 1: Check for null values. If you find any, replace the dataframe omitting that null valued row.

```python
# check for null values
swdf["RespondentID"].isnull().sum()
```

```python
#replacing the null values
swdf = swdf[pd.notnull(swdf["RespondentID"])]
```

```python
#checking once again
swdf["RespondentID"].isnull().sum()
```

Step 2: Checking if the other columns also have the data in the required format.

```python
swdf['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
for col in swdf.columns:
    print(swdf[col].value_counts(), '\n')
```

As the data is not in the required fomat, we need to change it. In each step, I would reduce few columns data into desired format.

Step3: col 1 and col 2 has Yes/No values, so we can easily change into True/False.

```python
#crearting a dictionary which is used for mapping
yes_no = {"Yes":True,"No":False}
```

```python
for col in ['Have you seen any of the 6 films in the Star Wars franchise?',
        'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    swdf[col]= swdf[col].map(yes_no)
```

```python
print(swdf['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())
print(swdf['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
```

Step 4: For col 3 to 8, it is multiple choice question, **"Which of the following Star Wars films have you seen? Please select all that apply."** where the reviewer was asked if they had watched the episode.

So, I am renaming all the *unnamed* to `seen#` where # represents the number of the episode. `NaN` is an empty value. You need not worry about it.

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}
```

```python
for col in swdf.columns[3:9]:
    swdf[col]=swdf[col].map(movie_mapping)
```

```python
for col in swdf.columns[3:9]:
    print(swdf[col].value_counts())
```

```python
swdf = swdf.rename(columns = {
       'Which of the following Star Wars films have you seen? Please select all that apply.':"seen1",
       'Unnamed: 4':"seen2", 
       'Unnamed: 5':"seen3", 
       'Unnamed: 6':"seen4", 
       'Unnamed: 7':"seen5",
       'Unnamed: 8':"seen6"})
```

```python
swdf.head()
```

Step 5: For col 9 to 14, it is also multiple choice question, **"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film."** where the reviewer was asked to rank the episodes.

So, I am renaming all the *unnamed* to `rank#` where # represents the number of the episode. `NaN` is an empty value. You need not worry about it.

```python
swdf[swdf.columns[9:15]] = swdf[swdf.columns[9:15]].astype(float)
```

```python
swdf = swdf.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "rank1",
        "Unnamed: 10": "rank2",
        "Unnamed: 11": "rank3",
        "Unnamed: 12": "rank4",
        "Unnamed: 13": "rank5",
        "Unnamed: 14": "rank6"
        })
```

```python
swdf.head().transpose()
```

```python
Now that the data looks clean, lets start the analysis.
```

# Highest/Lowest ranked Star Wars Movie:

```python
swdf[swdf.columns[9:15]].mean()
```

Considering it was suggested that the respondants rank their favorite movies from 1-5 where 1 would be their favorite movie and 5 will be their least favorite movie it seems like ranking_5 or the movie "Star Wars : The Empire Strikes Back" as the best rated so far with it's rating being 2.5 and the movie_3 i.e "Revenge of the sith" was the lowest rated movie here.

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

index = ["Star Wars: Episode I  The Phantom Menace",
    "Star Wars: Episode II  Attack of the Clones",
    "Star Wars: Episode III  Revenge of the Sith",
    "Star Wars: Episode IV  A New Hope",
    "Star Wars: Episode V The Empire Strikes Back",
    "Star Wars: Episode VI Return of the Jedi"]

plt.bar(range(6), swdf[swdf.columns[9:15]].mean())
plt.xticks(range(6),index,rotation = "vertical")
plt.xlabel("Star Wars movies preference chart(1:most favorite, 5: least favorite)")
plt.ylabel("Rating from surveymonkey users")
```

Note : The last 3 movies were actually published earlier and are considered as "Original movies" which seems to be rated way higher than the first 3 prequels.

# Most/Least viewed Star Wars Movie:

# Gender Based Segmentation to understand Viewer Preference:

```python
males = swdf[swdf["Gender"] == "Male"]
females = swdf[swdf["Gender"] == "Female"]
```

```python
males[males.columns[9:15]].mean()
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.xticks(range(6),index,rotation = "vertical")
plt.xlabel("Star Wars Movies preference chart for Male Viewers(1:Most Favorite, 5:Least Favorite)")
plt.ylabel("Rating from Surveymonkey Users")
```

```python
females[females.columns[9:15]].mean()
```

```python
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.xticks(range(6),index,rotation = "vertical")
plt.xlabel("Star Wars Movies preference chart for Female Viewers(1:Most Favorite, 5:Least Favorite)")
plt.ylabel("Rating from Surveymonkey Users")
```

'Empire Strikes Back' was rated as the best movie both by male and female viewers, but it does seem like male users hated the prequels more, considering female users seemed to like the first movie of the prequel i.e 'Phantom Menace'. 'Revenge of the Sith' is hated by both male and female viewers, but personally I thought it was great!!

# Star Wars Movie rating breakdown plot by Gender:

```python
N = 6
menMeans = males[males.columns[9:15]].mean()


ind = np.arange(N)  # the x locations for the groups
width = 0.40       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, menMeans, width, color='r')

womenMeans = females[females.columns[9:15]].mean()
rects2 = ax.bar(ind + width, womenMeans, width, color='y')

#Add some text for labels, Title and Axes ticks
ax.set_ylabel('Star wars Movie rating by Gender(1:Most Favorite,5:Least Favorite)')
ax.set_xticks(ind + width)
ax.set_xticklabels(index, rotation = "vertical")

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))


plt.show()
```

# Star Wars Viewer count breakdown plot by Gender:

```python
N = 6
menMeans = males[males.columns[9:15]].sum()


ind = np.arange(N)  # the x locations for the groups
width = 0.40       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, menMeans, width, color='g')

womenMeans = females[females.columns[9:15]].sum()
rects2 = ax.bar(ind + width, womenMeans, width, color = '#000fff')

#add some text for labels, title and axes ticks
ax.set_ylabel('Star wars Viewer Counts')
ax.set_xticks(ind + width)
ax.set_xticklabels(index, rotation = "vertical")

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))


plt.show()
```
